```python
import os
import glob
import numpy as np
import matplotlib.pyplot as plt
```

```python
foldername = 'DAH_Complete_Sorting_5'
```

```python
filepathregex = foldername + '_persimg_serialize' + os.sep + '*.npy'

for filepath in glob.glob(filepathregex):
    
    filename = filepath.split(os.sep)[1]
    persimgdata = np.load(filepath, allow_pickle=True)
    persimgdict = persimgdata.item()
    green_img_dim0 = persimgdict['H0'][0]
    red_img_dim0 = persimgdict['H0'][1]
    all_img_dim0 = persimgdict['H0'][2]
    
    plt.figure(figsize=(8,3), dpi=100)
    plt.plot(green_img_dim0, 'g')
    plt.plot(red_img_dim0, 'r')
    plt.plot(all_img_dim0, 'k')
    plt.title("H0 Persistence Image")
```
